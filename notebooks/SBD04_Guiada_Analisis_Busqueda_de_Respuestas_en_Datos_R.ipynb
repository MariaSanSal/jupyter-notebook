```r
v <- 1:15
(mi_dataframe <- data.frame(
    x=letters[1:5],
    y=v,
    z=v>2
    ))
```

```r
mi_dataframe[2:3,c(1,2)]
```

```r
mi_dataframe2 <- data.frame(
    col4 = 6:10,
    col5 = v+3,
    col6 = v < 3
)

cbind(mi_dataframe, mi_dataframe2)
```

```r
mi_dataframe3 <- data.frame(
    x=letters[6:8],
    y= 6:8,
    z= c(TRUE,FALSE,TRUE)
)

rbind(mi_dataframe,mi_dataframe3)
```
