```python
import pandas as pd

indice = ['uno','dos','tres','cuatro','cinco']
columnas = ['Raza','Puntos']
lista = [['Caniche',8.1],['Bulldog',7.3],['ChowChow',7.6],['Chihuahua',9.0],['Labrador',9.3]]
```

```python

print('----------------------------')
print('--- Apartado 1            --')
print('----------------------------')

#Apartado 1: crear un dataframe df con la lista de razas, las columnas e índice
df=pd.DataFrame(lista,columns=columnas,index=indice)

print(df)
```

```python
print('----------------------------')
print('--- Apartado 2            --')
print('----------------------------')

#Añadir una nueva columna 'PuntosAna' con la siguiente información puntos_ana = [61,75,82,95,99]
puntos_ana = [61,75,82,95,99]
df['PuntosAna']=puntos_ana

print(df)
```

```python
print('----------------------------')
print('--- Apartado 3            --')
print('----------------------------')

#Dividir la columna 'PuntosAna' entre 10 de modo que todas las puntuaciones queden en la misma escala

df.PuntosAna=df.PuntosAna.apply(lambda x: x/10)

print(df)
```

```python
print('----------------------------')
print('--- Apartado 4            --')
print('----------------------------')

#Añadir nuevos registros al dataframe lista2 = [['Samoyedo',9.2,8.9],['Pinscher',8.1,6.7]]
#Como índices usar 'nuevo1' y 'nuevo2' 
lista2 = [['Samoyedo',9.2,8.9],['Pinscher',8.1,6.7]]
indice2=['nuevo1','nuevo2']
columnas = ['Raza','Puntos','PuntosAna']
df2=pd.DataFrame(lista2,columns=columnas,index=indice2)
#df=df.append(df2) #indica que está obsoleto con un warning que recomienda concat
df=pd.concat([df,df2])

print(df)
```

```python

print('----------------------------')
print('--- Apartado 5            --')
print('----------------------------')

#Crea la columna 'Media' para poder ver la puntuación media para cada raza.
df['Media']=(df.Puntos+df.PuntosAna)/2 

print(df)
```

```python

print('----------------------------')
print('--- Apartado 6            --')
print('----------------------------')
#Eliminar la fila cuyo índice es 'dos'
df=df.drop('dos')

print(df)
```

```python
print('----------------------------')
print('--- Apartado 7            --')
print('----------------------------')
#deciden que la decisión va a estar entre chihuahua, labrador y samoyedo, por lo que deciden obtener sólo esas filas del DataFrame
df=df[2:5] 

print(df)
```
